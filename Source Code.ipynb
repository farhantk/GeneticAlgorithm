```python
# Package yang digunakan
import random
import math
```

```python
# Fungsi menghitung fitness, semakin nilainya besar semakin baik
def fitness(x, y):

  h=(math.cos(x)+math.sin(y))**2 / ((x**2)+(y**2))
  if h==0:
    h += 0.00000000000000000000000000000001
  return 1/h
```

```python
def decode(kro):
  p =0
  q =0
  r =0
  for i in range(16):
    p += 2**(-i)
    q += kro[i]*(2**(-i))
    r += kro[16+i]*(2**(-i))

  x = rb + (( (ra-rb) /  (p) * (q)))
  y = rb + (( (ra-rb) /  (p) * (r)))
  return x,y
```

```python
# Membuat kromosom
def createKro() :
  kromosom = Kromosom()
  bit = []
  for i in range(32):
    temp = random.randint(0, 1)
    bit.append(temp)
  kromosom.kro = bit
  x,y = decode(kromosom.kro)
  kromosom.fitness = fitness(x,y)
  return kromosom
```

```python
# Membuat populasi sebanyak n
def createPopulasi(n):
  populasi = []
  for i in range(n) :
    populasi.append(createKro())
  return populasi
```

```python
# Single Point Crossover
def crossover(a, b):
  p = Kromosom()
  q = Kromosom()
  child=[]
  prob = random.uniform(0, 1)
  prob2 = random.uniform(0.85, 1)
  if prob <= prob2:
    for i in range(2):
      hua = random.randint(10, 23)
      kro = a.kro[:hua] + b.kro[hua:]
      kro1 = b.kro[:hua] + a.kro[hua:]

      p.kro = kro
      x,y = decode(p.kro)
      p.fitness = fitness(x,y)
      
      q.kro = kro1
      x,y = decode(q.kro)
      q.fitness = fitness(x,y)

      child.append(p)
      child.append(q)
  else:
    child.append(a)
    child.append(a)
    child.append(b)
    child.append(b)
  return child
```
